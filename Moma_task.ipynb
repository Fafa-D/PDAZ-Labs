```python
#a. Find 5 artists with the most artworks in the dataset.

import csv

#read the file 

with open('moma_artworks_1000.csv', mode ='r', encoding= 'utf-8')as file:
  csvFile = csv.reader(file)

#create list of artist names

  artist_names = []                                              
  for row in csvFile:
    artist_names.append(row[2])


#create dictionary counting artist names

  artist_count = {}
  for name in artist_names:
    if name not in artist_count:
      artist_count[name] = 1
    else:
      artist_count[name] = artist_count[name] + 1

  
#order dictionary from word with highest occurence using the key parameter 
    
  names_sorted = sorted(artist_count.items(), key=lambda item: item[1], reverse = True) 


#shorten the tuple to the first 5 (one contained 3 artists)

  names_numbers = names_sorted[0:3] 


#create list of just artist names

  top_five = []
  for x in names_numbers:
    top_five.append(x[0])

  print(f"The 5 artists with the most artworks in the dataset are {', '.join(top_five)}.")  


```
