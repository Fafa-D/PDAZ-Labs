```python
url_line_stations = "https://api.tfl.gov.uk/Line/Victoria/StopPoints"

response_ls = requests.get(url_line_stations)
data_ls = response_ls.json()

df_ls = pd.DataFrame(data_ls)

df_ls.head(10)
```

```python
#first index is a list of dictionaries for 1 station
print(df_ls["additionalProperties"][1])
```

```python
#second index is one dictionart ie. one facility for that station
print(df_ls["additionalProperties"][1][3])
```

```python
iter = (df_ls["additionalProperties"][0])          #list of dictionaries
for i in iter:
   
    # now i is a dict, now we see the keys 
    # of the dict
    for key in i.keys():
        if key == "key":
            print(key)
```
