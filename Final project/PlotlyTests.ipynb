```python
#plotting stops on victoria line
#could do line colours for different lines from guidelines, how do we deal with multiples?
#could I do a function that prints the same analysis for each line like this line has x stops. there are x toilets, x stops have wifi, x stops are step
#free etc. There are currently x lift disruptions on this line.

url_line_stations = "https://api.tfl.gov.uk/Line/Victoria/StopPoints"             #access the API

response_ls = requests.get(url_line_stations)                                    
data_ls = response_ls.json()

df_ls = pd.DataFrame(data_ls)                                                     #create data frame from the API

fig_ls = px.scatter_map(df_ls, lat="lat", lon="lon", hover_name="commonName", hover_data=["naptanId", "modes"], zoom=9.5)      
fig_ls.update_layout(title='Victoria line stations', hovermode='closest', map_style="light", hoverlabel_bgcolor='rgb(80, 133, 217)')    
fig_ls.update_traces(marker=dict(color='LightSkyBlue', size=10))
fig_ls.show()
```

```python
site_lat = df_ls.lat
site_lon = df_ls.lon

fig_ls2 = go.Figure()

fig_ls2.add_trace(go.Scattermap(
        lat=site_lat,
        lon=site_lon,
        mode='markers',
        hovertext=df_ls['naptanId'],
        hoverinfo="text",
        marker=go.scattermap.Marker(
            size=10,
            color='rgb(255, 0, 0)',
            opacity=0.7
        )
    ))

fig_ls2.update_layout(title='Victoria line stations', autosize=True, hovermode='closest')
fig_ls2.update_layout(map_zoom=9, map_center_lat=51.5072, map_center_lon=0, map_style="light", hoverlabel_bgcolor='rgb(80, 133, 217)')
#fig_ls2.update_traces(hovertemplate = "{hovertext} <br><br>" + "longitude: %{lon}<br>" + "latitude: %{lat}<br>" )

fig_ls2.show()
```
