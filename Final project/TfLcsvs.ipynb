### Toilets.csv 
Exploration and creating a new csv to merge with data for final project

```python
df_Toilets = pd.read_csv("Toilets.csv")   

#First 10 rows of the csv
df_Toilets.head(10)
```

```python
Toilets = df_Toilets[["StationUniqueId", "IsAccessible", "Type"]]

print(Toilets)
```

```python
#Count the number of toilets at each station in list

Toilet_dict_count = {}

for naptan in Toilets["StationUniqueId"]:
    if naptan not in Toilet_dict_count:
        Toilet_dict_count[naptan] = 1
    else:
        Toilet_dict_count[naptan] = Toilet_dict_count[naptan] + 1


print(f"This dictionary contains the number of toilets at {len(Toilet_dict_count)} stations.")
print(Toilet_dict_count)
```

```python
#create list of accessible toilets, making sure there are no duplicates

Toilet_list_access = []

for index, row in Toilets.iterrows():
    if row["IsAccessible"] == "TRUE" and row["StationUniqueId"] not in Toilet_list_access:
        Toilet_list_access.append(row["StationUniqueId"])


print(f"This is a list of {len(Toilet_list_access)} stations that contain accessible toilets.")
print(Toilet_list_access)
```

```python
#create dataframe from these
df_Toilet_count = pd.DataFrame.from_dict(Toilet_dict_count, orient='index')

df_Toilet_count = df_Toilet_count.reset_index()

df_Toilet_count
```

```python
df_Toilet_count["HasAccessible"] = df_Toilet_count["index"].isin(Toilet_list_access)

df_Toilet_count
```

```python
#checking that not all values are true
access_check = []

for x in df_Toilet_count["HasAccessible"]:
    access_check.append(x)

print(access_check)
```

```python
#checking for correct number of accessible toilets
access_check_true = []

for x in access_check:
    if x == True:
        access_check_true.append(x)

print(access_check_true)
print(len(access_check_true))
```

```python
#rename columns
df_Toilet_count.rename(columns={'index': 'naptanId', 0: 'NumToilets', 'HasAccessible': 'HasAccessibleToilets'}, inplace=True)

df_Toilet_count
                       
```

```python
#Turning data to csv to use in Final Project file

df_Toilet_count.to_csv('CustomTfLToilet.csv', index=False)
```
